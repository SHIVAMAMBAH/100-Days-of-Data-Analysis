# Trivariate Analysis

### Column: country & year & suicides_no

```python
df.groupby(["country","year"])[ "suicides_no"].sum().reset_index(name = "total number of suicides in that year")
```

### Columns: country & sex & suicides_no

```python
df.groupby(["country","sex"])["suicides_no"].sum().reset_index(name = "total number of suicides")
```

### Columns: country & year & population

```python
df.groupby(["country", "year"])["population"].mean().reset_index(name = "average population in that year")
```

### Columns: country & year & gdp_for_year ($)

```python
df.groupby(["country", "year"])["gdp_for_year ($)"].mean().reset_index(name = "average gdp for that year")
```

### Columns: country & year & gdp_per_capita ($)

```python
top_countries = df['country'].value_counts().head(5).index
filtered_df = df[df['country'].isin(top_countries)]

# Group and calculate
df_grouped = filtered_df.groupby(["country", "year"])["gdp_per_capita ($)"].mean().reset_index(name="average_gdp")

# Plot
plt.figure(figsize=(18, 8))
sns.lineplot(data=df_grouped, x="year", y="average_gdp", hue="country", marker="o")

plt.title("Average GDP per Capita Over Years by Country", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Average GDP per Capita ($)")
plt.legend(title="Country")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Columns: year & sex & suicides_no

```python
df_grouped = df.groupby(["year", "sex"])["suicides_no"].sum().reset_index(name = "total number of suicides by the gender in that year")

plt.figure(figsize=(20, 6))
sns.lineplot(data=df_grouped, x="year", y="total number of suicides by the gender in that year", hue="sex", marker="o")

plt.title("Average GDP per Capita Over Years by Gender", fontsize=14)
plt.xlabel("Year")
plt.ylabel("total number of suicides by the gender in that year")
plt.legend(title="Sex")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Columns: year & sex & gdp_per_capita ($)

```python
df_grouped = df.groupby(["year", "sex"])["gdp_per_capita ($)"].mean().reset_index(name = "average gdp per capita of the gender in that year")

plt.figure(figsize=(20, 6))
sns.lineplot(data=df_grouped, x="year", y="average gdp per capita of the gender in that year", hue="sex", marker="o")

plt.title("Average GDP per Capita Over Years by Gender", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Average GDP per Capita ($)")
plt.legend(title="Sex")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Tetravariate Analysis

### Columns: country, year, sex & suicides_no

```python
display(df.groupby(["country", "year", "sex"])["suicides_no"].sum().reset_index(name = "total number of suicides by the gender in that year for that country"))

# Step 1: Group the data
df_grouped = df.groupby(["country", "year", "sex"])["suicides_no"].sum().reset_index(
    name="total_suicides"
)

# Step 2: Limit to top 3 or 4 countries for clarity (optional)
top_countries = df['country'].value_counts().head(8).index
filtered_df = df_grouped[df_grouped['country'].isin(top_countries)]

# Step 3: Plot
plt.figure(figsize=(20, 10))
sns.lineplot(
    data=filtered_df,
    x="year",
    y="total_suicides",
    hue="sex",
    style="country",
    markers=True,
    dashes=False
)

plt.title("Total Suicides per Year by Gender in Selected Countries", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Total Number of Suicides")
plt.legend(title="Gender / Country")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Columns: country, year, sex & gdp_per_capita ($)

```python
display(df.groupby(["country", "year", "sex"])["gdp_per_capita ($)"].mean().reset_index(name = "average gdp per capita of the gender in that year for that country"))

df_grouped = df.groupby(["country", "year", "sex"])["gdp_per_capita ($)"] \
               .mean() \
               .reset_index(name="average_gdp")


top_countries = df['country'].value_counts().head(20).index
filtered_df = df_grouped[df_grouped['country'].isin(top_countries)]

plt.figure(figsize=(20, 10))
sns.lineplot(
    data=filtered_df,
    x="year",
    y="average_gdp",
    hue="sex",
    style="country",  # different line styles per country
    markers=True,
    dashes=False
)

plt.title("Average GDP per Capita by Gender Over Years (Selected Countries)", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Average GDP per Capita ($)")
plt.legend(title="Gender / Country")
plt.grid(True)
plt.tight_layout()
plt.show()
```
