# Univariate Analysis
Analyzing one column (variable) at a time to find its:

- Central tendency
- Spread of Data
- Shape of Distribution
- Visual Representation

We will analyze each column `one by one`

### Column: Song_Title

```python
print("*-------------------Basic Info About the Column (Song_Title)-------------------*")
print(f"Number of rows in the column: {df['Song_Title'].shape[0]}")
print(f"Number of NULL/NaN values: {df['Song_Title'].isnull().sum()}")
print(f"Type of the Column: {df['Song_Title'].dtype}")
print(f"Number of Unique Values in the Column: {df['Song_Title'].nunique()}")
print(f"Memory Usage of the Column: {df['Song_Title'].memory_usage(deep = True) / (1024 ** 1)} KB")
print(f"Percentage of total memory occupied by COlumn: {((df['Song_Title'].memory_usage(deep = True) / (1024 ** 1))/(df.memory_usage(deep = True).sum() / (1024 ** 1)))*100}")

print("\n")

print("*-------------------Statistical summary of the Column (Song_Title)-------------------*")
print(df["Song_Title"].describe())

print("\n")

print("*-------------------Graphical Visualization of the Column (Song_Title)-------------------*\n")
plt.figure(figsize = (8, 6), dpi = 100, facecolor = "white", edgecolor = "black")
plt.bar(
    df["Song_Title"].value_counts().index, df["Song_Title"].value_counts().values,
    color = "orange",
    align = "center",
    width = 0.6,
    edgecolor = "black",
    linewidth = 1,
    zorder = 3,
)
plt.title("Bar Plot of Song_Title", fontsize = 20, fontweight = "bold", loc = "center")
plt.xlabel("Song Title",fontsize = 14, fontweight = "bold", color = "black")
plt.ylabel("Count",fontsize = 14, fontweight = "bold", color = "black")
plt.xticks(rotation = 90, fontsize = 12)
plt.yticks(rotation = 0, fontsize = 12)
plt.tight_layout()
plt.show()
```

# Bivariate Analyis
Bivariate analysis is a statistical method that examines the relationship between two variables. The goal is to determine whether and how the variables are relatedwhether changes in one variable are associated with changes in the other.
- Artist & Song_Title
- Genre & Song_Title
- Artist & Duration_Minutes
- Genre & Duration_Minutes
- Platform & Song_Tile
- Platform & Artist
- Release_Year & Song_Title

### Columns: Artist & Song_Title

```python
display(df.groupby("Artist")["Song_Title"].value_counts().reset_index(name = "No. of times song released"))
```

### Columns: Genre & Song_Title

```python
display(df.groupby("Genre")["Song_Title"].value_counts().reset_index(name = "No. of times song released"))
```

### Column: Artist & Duration_Minutes

```python
display(df.groupby("Artist")["Duration_Minutes"].mean().reset_index(name = "Average Timing of Song"))
```

### Column: Genre & Duration_Minutes

```python
display(df.groupby("Genre")["Duration_Minutes"].mean().reset_index(name = "Average Timing of Song"))
```

### Colums: Platform & Song_Title

```python
display(df.groupby("Platform")["Song_Title"].value_counts().reset_index(name = "No. of times song released"))
```

### Column: Platform & Artist

```python
display(df.groupby("Platform")["Artist"].value_counts().reset_index(name = "No. of times Artist's song released"))
```

### Column: Release_Year & Song_Title

```python
display(df.groupby("Release_Year")["Song_Title"].value_counts().reset_index(name = "No. of times song released in that year"))
```
